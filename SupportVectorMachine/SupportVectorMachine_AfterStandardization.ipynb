```python
import pandas as pd

dataset = pd.read_csv("cars_dataset.csv")

dataset.columns
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['buying_low', 'buying_med', 'buying_vhigh', 'maint_low', 'maint_med',
       'maint_vhigh', 'doors_four', 'doors_three', 'doors_two', 'persons_more',
       'persons_two', 'lug_boot_med', 'lug_boot_small', 'safety_low',
       'safety_med']]
```

```python
dependent = dataset[[ 'car_vgood']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.30, random_state=0) 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.svm import SVR
```

```python
#regressor = SVR(kernel = "linear")
#regressor = SVR(kernel = "rbf")
regressor = SVR(kernel = "poly") # https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
```

```python
regressor.fit(x_train, y_train)
```

```python
# weight = regressor.coef_ # view the wt parameter for linear kernel
weight = regressor.support_ # view the support vectors for rbf kernel
```

```python
weight
```

```python
bias = regressor.intercept_
```

```python
bias
```

```python
y_predict = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
error = r2_score(y_predict, y_test)
```

```python
error
```
